Diffusion is autoregression in the frequency domain？

[EEML'24] Sander Dieleman - Generative modelling through iterative refinement      
“diffusion is(approximate) spectral autoregressive”  [42:55](https://www.youtube.com/watch?v=9BHQvQlsVdE&list=WL&index=3&t=11s)

```python
from matplotlib import pyplot as plt
import cv2
import numpy as np
img = cv2.imread('./images/n02894605_breakwater.jfif', cv2.IMREAD_COLOR)
# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# scale to [-1, 1]
def rescale(x):
    range_ = np.ptp(x)
    min_ = np.min(x)
    # -> [0, 1]
    x = (x - min_) / range_
    # -> [0, 2]
    # ->  [-1 , 1]
    return 2*x - 1
img = rescale(img)
print(img.shape)
plt.imshow(img, cmap='gray', vmin=-1, vmax=1)
plt.title("orgin")
plt.show()
```

```python
def gen_gaussian_noise(mu, sigma, n=img.shape):
    return np.random.normal(mu, sigma, n)

noise_0_1 = gen_gaussian_noise(0, 1)
# print(noise_2d)
plt.imshow(noise_0_1, cmap='gray', vmin=-1, vmax=1)
plt.title("noise_0_1")
plt.show()
```

```python
img_add_noise_0_1 = img + noise_0_1
img_add_noise_0_1 = rescale(img_add_noise_0_1)
plt.imshow(img_add_noise_0_1, cmap='gray', vmin=-1, vmax=1)
plt.title("img_add_noise_0_1")
plt.show()
```

```python
sigmas = [x / 10.0 for x in range(20)]
print(sigmas)
```

```python
for sigma in sigmas:
    noise = gen_gaussian_noise(0, sigma)
    image_add_noise = img + noise 
    image_add_noise = rescale(image_add_noise)
    plt.imshow(img_add_noise_0_1, cmap='gray', vmin=-1, vmax=1)
    plt.title(f"sigma{sigma}")
    plt.show()
    
    
```
